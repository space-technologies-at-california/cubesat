```python
import numpy as np
import scipy.special as special
import matplotlib.pyplot as plt
```

```python
snr = np.linspace(0,20,50)
ber = 1/2*special.erfc(1/2*np.sqrt(snr))
plt.plot(snr, np.log(ber))
plt.show()
```

```python
background = np.power(10,-17.4) #sqrt(V)/Hertz
bandwidth = 1e5 #Hertz
noise_power = background*bandwidth #sqrt(V)
required_snr = 5.0 #dB
transmit_power = .0001 #Watts
transmit_eff = .6
comm_distance = float(input("Enter desired distance from PCB to CubeSat for a Comm Link: ")) #meters
reciever_radius = .04 #meters
recieve_eff =.6
beam_divergence = float(input("Enter desired beam divergence (in degrees): ")) #degrees
beam_divergence_rads = beam_divergence*np.pi/180
```

```python
beam_divergence_rads = np.linspace(0.0001,.1*np.pi, 100)
recieve_pow = (transmit_power*transmit_eff*recieve_eff*(reciever_radius**2))/(beam_divergence_rads**2*comm_distance**2)
real_snr = recieve_pow/noise_power
plt.plot(beam_divergence_rads/np.pi*180, np.log10(real_snr))
plt.plot(beam_divergence_rads/np.pi*180, np.ones_like(beam_divergence_rads/np.pi*180)*np.log10(required_snr))
plt.show()
```
